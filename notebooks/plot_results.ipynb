```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error
```

```python
data = np.loadtxt("/home/rhys/PhD/sampnn/test_results.csv", delimiter=",")

plt.figure(figsize=(7,7))
plt.plot(data[:,1], data[:,2],'x')
plt.plot((np.min(data[:,1:]),np.max(data[:,1:])), (np.min(data[:,1:]),np.max(data[:,1:])), 'r--')
plt.show()

print("R2 Score: {:.4f}     [target 0.90]".format(r2_score(data[:,1],data[:,2])))
print("RMSE:     {:.4f} eV  [target 0.45 eV]".format(np.sqrt(mean_squared_error(data[:,1],data[:,2]))))
```
